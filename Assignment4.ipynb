```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

df_income = pd.read_csv('per_capita_income.csv')

for i in range(0,10):
    df_income.ix[i , 'Year'] = 2005+i 
 

df_income = df_income[['Year', 'Comparative Figures (Rupees.) All-India At Constant Prices']]


df_income = df_income.set_index('Year')
#taking common years
df_income1 = df_income[5:]
df_income1
x_vals1 = list(df_income1.index)
y_vals1 = df_income1['Comparative Figures (Rupees.) All-India At Constant Prices']
```

```python
df_edu = pd.read_csv('Gross_higher_education_enrollment.csv')
df_edu = df_edu.set_index('Country/ State/ UT Name')
df_all_india = df_edu.loc['All India']
df_all_india

for i in range(0,6):
    df_all_india.ix[i , 'Year'] = 2010+i

df_all_india1 = df_all_india[0:5]    

x_vals2 = list(df_all_india1['Year'])
y_vals2 = df_all_india1['All Categories - Total']

#plt.plot( x_vals2 , y_vals2 , 'r')
#plt.show()
```

```python
import seaborn as sns

sns.set_style('ticks')

fig = plt.figure()
ax =  fig.add_subplot(111)
plt.grid(color = 'grey' , alpha = 0.5 , axis= 'y', linestyle = '--')
plt.plot(x_vals1 , y_vals1 , 'g-o')
  

plt.ylabel('Per Capita Income' , color = 'g')
plt.title('Per Capita Income vs Higher Education (2010 - 2015)' , size = 12 , alpha = 0.99)
plt.legend(['Per Capita Income'] , frameon = False , bbox_to_anchor=(0.45, 0.3), loc=2)
plt.xticks(rotation = 30)
plt.tick_params('y' , bottom = 'off' , left = 'off' , colors='g')
plt.tick_params('x' , bottom = 'off' , left = 'off' )

ax2 = ax.twinx()
ax2.plot( x_vals2 , y_vals2, 'r-o')
ax2.set_ylabel('Gross Enrollment Ratio(GER)', color='r')
plt.legend(['Higher Education Enrollment'] , frameon = False , bbox_to_anchor=(0.45, 0.2), loc=2)
ax2.tick_params( 'y' , right = 'off', colors='r')
plt.tick_params('x' , bottom = 'off' , left = 'off')
plt.xticks(rotation = 30)
for spine in ax.spines.values():
    spine.set_visible(False)

for spine in ax2.spines.values():
    spine.set_visible(False) 



labels = [item.get_text() for item in ax.get_xticklabels()]
labels[0] = 'Jan 2010'
labels[1] = 'Jul 2010'
labels[2] = 'Jan 2011'
labels[3] = 'Jul 2011'
labels[4] = 'Jan 2012'
labels[5] = 'Jul 2012'
labels[6] = 'Jan 2013'
labels[7] = 'Jul 2013'
labels[8] = 'Jan 2014'
labels[9] = 'Jul 2014'
labels[10] = 'Jan 2015'
ax.set_xticklabels(labels)
ax2.set_xticklabels(labels)
plt.xticks(rotation = 30)


plt.show()
```
